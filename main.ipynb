```python
# detecting the pitch
def pitch_detection(data, sample_rate):
    frame_size = 2048
    hop_size = 512
    min_freq = 80
    max_freq = 1000
    
    times = []
    pitches = []
    
    min_lag = int(sample_rate/max_freq)
    max_lag = int(sample_rate/min_freq)

    for start in range(0, len(data)-frame_size, hop_size):
        frame = data[start:start+frame_size]
        
        # removing dc offset
        frame = frame-np.mean(frame) 
        
        # Autocorrelation
        corr = np.correlate(frame, frame, mode="full")
        corr = corr[frame_size:]
        
        # find peak in valid lag range
        peak_index = np.argmax(corr[min_lag:max_lag])+min_lag
        peak_value = corr[peak_index]
        
        if peak_value>0:
            pitch = sample_rate/peak_index
        else:
            pitch = 0
        
        time = start/sample_rate
        times.append(time)
        pitches.append(pitch)
    
    return times, pitches
        

times, pitches = pitch_detection(data_mono, sample_rate)

print(times)
print(pitches)
```

```python
note_names = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']

def pitch_to_notes(pitch):
    if pitch == 0:
        return "X"
    
    A4_freq = 440.0
    
    semitone_offset = 12 * np.log2(pitch/A4_freq)
    nearest_semitone = round(semitone_offset)
    
    idx_note = (nearest_semitone+9)%12
    octave = ((nearest_semitone+9)//12) + 4
    
    note = note_names[idx_note]
    return f"{note}{octave}"


notes = []

for pitch in pitches:
    note = pitch_to_notes(pitch)
    notes.append(note)

print(notes)
print(times)
```

```python

def note_to_vector(note):
    vec = np.zeros(12)
    if note == "X":
        return vec
    pitch_class = ''.join([c for c in note if c.isalpha() or c == '#'])
    if pitch_class in note_names:
        idx = note_names.index(pitch_class)
        vec[idx] = 1
    return vec
```

```python
# Shukla Ji's Part
def custom_dtw(cost_matrix):
    n, m = cost_matrix.shape
    D = np.full((n+1, m+1), np.inf)
    D[0, 0] = 0

    for i in range(1, n+1):
        for j in range(1, m+1):
            cost = cost_matrix[i-1, j-1]
            D[i, j] = cost + min(D[i-1, j],    # insertion
                                 D[i, j-1],    # deletion
                                 D[i-1, j-1])  # match

    # Backtrack to find the warping path
    i, j = n, m
    path = []
    while i > 0 and j > 0:
        path.append((i-1, j-1))
        steps = [D[i-1, j-1], D[i-1, j], D[i, j-1]]
        min_step = np.argmin(steps)
        if min_step == 0:
            i, j = i-1, j-1
        elif min_step == 1:
            i -= 1
        else:
            j -= 1
    path.reverse()
    return D[1:, 1:], path


def compare_audio_files(path1, path2):
    sr1, data1 = wavfile.read(path1)
    sr2, data2 = wavfile.read(path2)

    mono1 = data1.mean(axis=1) if data1.ndim > 1 else data1
    mono2 = data2.mean(axis=1) if data2.ndim > 1 else data2

    times1, pitches1 = pitch_detection(mono1, sr1)
    times2, pitches2 = pitch_detection(mono2, sr2)

    notes1 = [pitch_to_notes(p) for p in pitches1]
    notes2 = [pitch_to_notes(p) for p in pitches2]

    vecs1 = np.array([note_to_vector(n) for n in notes1])
    vecs2 = np.array([note_to_vector(n) for n in notes2])

    cost_matrix = cdist(vecs1, vecs2, metric='cosine')
    D, wp = custom_dtw(cost_matrix)

    sim_score = 1 - D[-1, -1] / len(wp)
    print(f"\n Comparing: {os.path.basename(path1)} vs {os.path.basename(path2)}")
    print(f" Similarity Score: {sim_score:.3f}")
    plot_alignment_line_graph(len(notes1), len(notes2), wp)
    return sim_score

def plot_alignment_line_graph(len1, len2, wp):
    x = [i for i, _ in wp]
    y = [j for _, j in wp]
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, marker='o', linestyle='-', color='blue')
    plt.title("Note Alignment Path (DTW Line Graph)")
    plt.xlabel("Time Index - Audio 1")
    plt.ylabel("Time Index - Audio 2")
    plt.grid(True)
    plt.xlim(0, len1)
    plt.ylim(0, len2)
    plt.tight_layout()
    plt.show()
path1 = "./audio1.wav"
path2 = "./audio2.wav"
compare_audio_files(path1,path2)
```
